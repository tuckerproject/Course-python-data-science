Next, we load the GloVe vectors.

```python
import numpy as np

embeddings_index = {}
f = open('glove.6B.100d.txt')
for line in f:
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:], dtype='float32')
    embeddings_index[word] = coefs
f.close()

print('Found %s word vectors.' % len(embeddings_index))
```

```python
jobs = ["secretary", "typist", "phlebotomist", "telemarketer", "packer", 
        "appraiser", "dentist", "rancher", "announcer", "firefighter"]   
femrates = [94.0,  85.1,  75.0,  65.3,  54.5,  45.3,  35.7,  25.8,  17.7,  5.1]

Aset = ["female", "woman", "girl", "sister", "she", "her", "hers", "daughter"]
Bset = ["male", "man", "boy", "brother", "he", "him", "his", "son"]

def cos(a,b):
  return((embeddings_index[a] @ embeddings_index[b])
         / np.sqrt(embeddings_index[a] @ embeddings_index[a]) 
         / np.sqrt(embeddings_index[b] @ embeddings_index[b]))

def s(A,B,w):
  return((np.mean([cos(w,a) for a in A])-np.mean([cos(w,b) for b in B]))
          / np.std([cos(w,x) for x in A+B]))
```

```python
#  50 .29
# 100 .29
# 200 .44
# 300 .41


np.corrcoef([s(Aset,Bset,w) for w in jobs],femrates)[0,1]
```
